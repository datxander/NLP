```python
# Slicing & dicing with HF Datasets

# Get the Drug review dataset from UC Irvine

!wget "https://archive.ics.uci.edu/ml/machine-learning-databases/00462/drugsCom_raw.zip"
!unzip drugsCom_raw.zip
```

```python
!pip install datasets
from datasets import load_dataset

data_files = {"train" : "drugsComTrain_raw.tsv", "test": "drugsComTest_raw.tsv"}
drug_dataset = load_dataset("csv", data_files = data_files, delimiter = "\t")
```

```python
drug_sample = drug_dataset["train"].shuffle(seed = 49).select(range(100))
drug_sample[:3]
```

```python
#Check if the Unnamed column could be the Patient ID

for split in drug_dataset.keys():
  assert len(drug_dataset[split]) == len(drug_dataset[split].unique("Unnamed: 0"))
```

```python
#Hypothesis confirmed. Let's use the DatasetDict.rename_column() to rename this column across both splits in 1 go

drug_dataset = drug_dataset.rename_column(
    original_column_name = "Unnamed: 0", new_column_name = "patient_id"
)

drug_dataset
```

```python
# Unique drug names
print("Unique drug count : ")
len(drug_dataset["train"].unique("drugName"))

# Unique condition names
print("Unique condition count : ")
len(drug_dataset["train"].unique("condition"))
```

```python
# Normalize the condition labels to lowercase

def lowercase_condition(example):
  return {"condition" : example["condition"].lower()}

drug_dataset.map(lowercase_condition)

```

```python
# That error came in because of some entries where condition = None. Let's exclude those with a lambda function

drug_dataset = drug_dataset.filter(lambda x: x["condition"] is not None)
```

```python
# Now we can normalise the condition field


def lowercase_condition(example):
  return {"condition" : example["condition"].lower()}

drug_dataset.map(lowercase_condition)
```

```python
# Dealing with the reviews column

# Count the number of words in each review

def compute_review_length(example):
  return {"review_length" : len(example["review"].split())}


```

```python
drug_dataset = drug_dataset.map(compute_review_length)
drug_dataset
```

```python
# Test the new dataset
drug_dataset["train"][20]
```

```python
# Check out the shortest 3 reviews

drug_dataset["train"].sort("review_length")[:3]
```

```python
# We want to predict the condition from the reviews, so very short reviews like these are not useful.
#Lets remove reviews with < 30 words

drug_dataset = drug_dataset.filter(lambda x: x["review_length"] > 30)
print(drug_dataset.num_rows)
```

```python
# From the reviewes, unscramble the HTML characters with the html library and the unescape function

import html
drug_dataset = drug_dataset.map(lambda x: {"review" : html.unescape(x["review"])})
```

```python
# Use batched = True for faster performance
import html
new_drug_dataset = drug_dataset.map(lambda x: {"review" : html.unescape(x["review"])}, batched = True)
```

```python
# We can extend this method to speed up the tokenizing

from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

def tokenize_function(examples):
  return tokenizer(examples["review"], truncation = True)
```

```python
# Time the below instruction using %time

%time tokenized_dataset = drug_dataset.map(tokenize_function, batched = True)
```

```python
slow_tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased", use_fast = False)

def slow_tokenize_function(examples):
  return slow_tokenizer(examples["review"], truncation = True)

tokenized_dataset = drug_dataset.map(slow_tokenize_function, batched = True, num_proc = 8)
```

```python
# We can return all chunks of text instead of the 1st one only, with the return_overflowing_tokens option

def tokenize_and_split(examples):
  return tokenizer(
      examples["review"],
      truncation = True,
      max_length = 128,
      return_overflowing_tokens = True
  )
```

```python
# Testing with one example

result = tokenize_and_split(drug_dataset["train"][1])
[len (ip) for ip in result["input_ids"]]
```

```python
# Expanding to the entire dataset

tokenized_dataset = drug_dataset.map(tokenize_and_split, batched = True)
```

```python
# Remove the columns from the old dataset

tokenized_dataset = drug_dataset.map(tokenize_and_split,
                                     batched = True,
                                     remove_columns = drug_dataset["train"].column_names)
```

```python
# The new datasets have many more elements

len(tokenized_dataset["train"]), len(drug_dataset["train"])
```

```python
# Conversion between libraries with set_format() option

drug_dataset.set_format("pandas")
```

```python
# The below output now comes as a dataframe

drug_dataset["train"][:5]
```

```python
# Let's convert the whole training dataset into a Pandas dataframe

train_df = drug_dataset["train"][:]
```

```python
train_df.head()
```

```python
# Now we can use pandas functionality like chaining

frequencies = (
    train_df["condition"].
    value_counts().
    to_frame().
    reset_index().
    rename(columns = {"index" : "condition", "condition" : "frequency"})
)

frequencies.head()
```

```python
# Convert this to a Dataset object

from datasets import Dataset

freq_dataset = Dataset.from_pandas(frequencies)
freq_dataset
```

```python
# Let's build a validation dataset

drug_dataset.reset_format()

drug_dataset_clean = drug_dataset["train"].train_test_split(train_size = 0.8, seed = 28)

# Rename the default "test" split to "validation"
drug_dataset_clean["validation"] = drug_dataset_clean.pop("test")

# Add this "test" dataset to the DatasetDict
drug_dataset_clean["test"] = drug_dataset["test"]

drug_dataset_clean
```

```python
# Let's save this dataset in the Arrow format

drug_dataset_clean.save_to_disk("drug-reviews")
```

```python
# once this is saved, it can be reloaded back from

from datasets import load_from_disk

drug_dataset_reloaded = load_from_disk("drug-reviews")
drug_dataset_reloaded
```
