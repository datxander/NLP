```python
# Slicing & dicing with HF Datasets

# Get the Drug review dataset from UC Irvine

!wget "https://archive.ics.uci.edu/ml/machine-learning-databases/00462/drugsCom_raw.zip"
!unzip drugsCom_raw.zip
```

```python
!pip install datasets
from datasets import load_dataset

data_files = {"train" : "drugsComTrain_raw.tsv", "test": "drugsComTest_raw.tsv"}
drug_dataset = load_dataset("csv", data_files = data_files, delimiter = "\t")
```

```python
drug_sample = drug_dataset["train"].shuffle(seed = 28).select(range(100))
drug_sample[:7]
```

```python
#Check if the Unnamed column could be the Patient ID

for split in drug_dataset.keys():
  assert len(drug_dataset[split]) == len(drug_dataset[split].unique("Unnamed: 0"))
```

```python
#Hypothesis confirmed. Let's use the DatasetDict.rename_column() to rename this column across both splits in 1 go

drug_dataset = drug_dataset.rename_column(
    original_column_name = "Unnamed: 0", new_column_name = "patient_id"
)

drug_dataset
```

```python
# Unique drug names
print("Unique drug count : ")
len(drug_dataset["train"].unique("drugName"))

# Unique condition names
print("Unique condition count : ")
len(drug_dataset["train"].unique("condition"))
```

```python
# Normalize the condition labels to lowercase

def lowercase_condition(example):
  return {"condition" : example["condition"].lower()}

drug_dataset.map(lowercase_condition)

```

```python
# That error came in because of some entries where condition = None. Let's exclude those with a lambda function

drug_dataset = drug_dataset.filter(lambda x: x["condition"] is not None)
```

```python
# Now we can normalise the condition field


def lowercase_condition(example):
  return {"condition" : example["condition"].lower()}

drug_dataset.map(lowercase_condition)
```

```python
# Dealing with the reviews column

# Count the number of words in each review

def compute_review_length(example):
  return {"review_length" : len(example["review"].split())}


```

```python
drug_dataset = drug_dataset.map(compute_review_length)
drug_dataset
```

```python
# Test the new dataset
drug_dataset["train"][15]
```

```python
# Check out the shortest 3 reviews

drug_dataset["train"].sort("review_length")[:3]
```

```python
# Lets remove reviews with < 30 words

drug_dataset = drug_dataset.filter(lambda x: x["review_length"] > 30)
print(drug_dataset.num_rows)
```

```python
# From the reviewes, unscramble the HTML characters with the html library and the unescape function

import html
drug_dataset = drug_dataset.map(lambda x: {"review" : html.unescape(x["review"])})
```
